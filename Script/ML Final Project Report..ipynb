### PROCESSING THE WHOLE DATA:
  #### 1) For train data and converting to a Numpy array:

```python
data = []

for category in CATEGORIES:
    path = os.path.join(Train_DIRECTORY, category) # path of dataset
    for img in os.listdir(path):
        try:
            img_path = os.path.join(path, img) #Getting the image path
            label = CATEGORIES.index(category)# Assigning label to image
            arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) # Converting image to grey scale
            new_arr = cv2.resize(arr, (100, 100)) # Resize image
            data.append([new_arr, label]) # appedning image and label in list
        except Exception as e:
            print(str(e))
```

```python
x_train = []
y_train = []
```

```python
for features, label in data:
    x_train.append(features)  # Storing Images all images in X
    y_train.append(label) # Storing al image label in y
```

```python
x_train = np.array(x_train) # Converting it into Numpy Array
y_train = np.array(y_train)
```

#### 2) For test data and converting to a Numpy array:

```python
data = []

for category in CATEGORIES:
    path = os.path.join(Test_DIRECTORY, category) # path of dataset
    for img in os.listdir(path):
        try:
            img_path = os.path.join(path, img) #Getting the image path
            label = CATEGORIES.index(category)# Assigning label to image
            arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) # Converting image to grey scale
            new_arr = cv2.resize(arr, (100, 100)) # Resize image
            data.append([new_arr, label]) # appedning image and label in list
        except Exception as e:
            print(str(e))
```

```python
x_test=[]
y_test=[]
for features, label in data:
    x_test.append(features)  # Storing Images all images in X
    y_test.append(label) # Storing al image label in y
```

```python
x_test = np.array(x_test) # Converting it into Numpy Array

y_test = np.array(y_test)
```

## **CNN MODEL.**

```python
import pickle
```

```python
x_train= pickle.load(open('x_train.pkl', 'rb')) # loading preproccesed images
y_train = pickle.load(open('y_train.pkl', 'rb'))# loading Labels

x_test=pickle.load(open('x_test.pkl', 'rb')) # loading preproccesed images
y_test = pickle.load(open('y_test.pkl', 'rb'))# loading Labels
```

### NORMALIZING IMAGES:
- pixel values in range 0 to 255 are normalized by dividing wih 255. 

```python
x_train= x_train/255 # normalizing images 
```

```python
x_test= x_test/255 
```

```python
x_train = x_train.reshape(-1, 100, 100, 1) # reshaping it for CNN
x_test = x_test.reshape(-1, 100, 100, 1) # reshaping it for CNN
```

### DESIGNING THE CNN MODEL:
- Convolutional Layer 1 has 64 filters of size 3 by 3 with an activation function of RELU and a maxpooling of 2 by 2 to reduce computation.
- Convolutional Layer 2 has 32 filters of size 3 by 3 with an activation function of RELU and a maxpooling of 2 by 2 to reduce computation.
- Flatten the layer output to feed into the ANN.
- Hidden Layer 1 has 30 neurons with the activation function RELU and a dropout layer to prevent overfitting.
- Hidden Layer 2 has 128 neurons with the activation function RELU and a dropout layer to prevent overfitting.
- Hidden Layer 3 has 64 neurons with the activation function RELU and a dropout layer to prevent overfitting.
- Output Layer has 2 neurons for two classes using Softmax activation function.

```python
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D,Activation,MaxPooling2D
```

```python
# Designing CNN
model = Sequential()

model.add(Conv2D(64, (3,3), activation = 'relu')) 
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(32, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(30, input_shape = x_train.shape[1:], activation = 'relu'))
model.add(Dropout(0.5)) ## precuation for overfitting 

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(2, activation = 'softmax')) ## output layer
```

- Optimizer is adam with lose function sparse_categorical_crossentropy as softmax is used in output Layer.

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### MODEL FIT:
- Epoch is 10.
- Validation split = 10%

```python
from matplotlib import pyplot as plt
history = model.fit(x_train, y_train,validation_split = 0.1, epochs=10)
```

### PERFORMANCE PLOT:

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

##  **TRANSFER LEARNING USING THE PRE-TRAINED VGG16 MODEL.**

### INTRODUCTION

The full name of VGG is the Visual Geometry Group, which belongs to the Department of Science and Engineering at Oxford University. It has released a series of convolutional network models beginning with VGG, which can be applied to face recognition and image classification, from VGG16 to VGG19. The original purpose of VGG's research on the depth of convolutional networks was to understand how the depth of convolutional networks affects the accuracy of large-scale image classification and recognition. -Deep-16 CNN) In order to deepen the number of network layers and avoid having too many parameters, a small 3x3 convolution kernel is used in all layers.
- Source Google.

### THE NETWORK STRUCTURE

The input of VGG is set to an RGB image of size 224x224. The average RGB value is calculated for all images on the training set image, and then the image is input as an input to the VGG convolution network. A 3x3 or 1x1 filter is used, and the convolution step is fixed. There are three VGG fully connected layers, which can vary from VGG11 to VGG19 according to the total number of convolutional layers plus fully connected layers. The minimum VGG11 has eight convolutional layers and three fully connected layers. The maximum VGG-19 has 16 convolutional layers. +3 completely interconnected layers In addition, the VGG network is not followed by a pooling layer behind each convolutional layer, for a total of five pooling layers distributed under different convolutional layers.
- Source Google.

### MODIFICATIONS IN THE PRE-TRAINED MODEL:
- Retain pre-trained weights by freezing layers.
- Modified input and output layers according to the image and target classes. 

```python
from keras.layers import Input, Lambda, Dense, Flatten
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
IMAGE_SIZE = [100, 100]
```

```python
#Give dataset path
train_path = 'dataset\train'
#test_path = 'data/test'
```

```python
vgg = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in vgg.layers:
  layer.trainable = False
```

```python
x = Flatten()(vgg.output)
prediction = Dense(2, activation='softmax')(x)
model = Model(inputs=vgg.input, outputs=prediction)
model.summary()
```

- loss function is binary_crossentropy and optimizer is SGD.

```python
import tensorflow as tf

sgd = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
model.compile(loss='binary_crossentropy',
              optimizer=sgd,
              metrics=['accuracy'])
```

- Data augumentation to create more training examples:

```python
# Data Augmentation z##
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
test_datagen = ImageDataGenerator()
```

```python
# provide the same target size as initialied for the image size
train_set = train_datagen.flow_from_directory(r'dataset\train',
                                                 target_size = (100, 100),
                                                 class_mode = 'categorical',batch_size = 32)
```

```python
test_set = test_datagen.flow_from_directory('dataset/val',
                                         target_size = (100, 100),
                                           class_mode = 'categorical',batch_size = 32)
```

```python
from datetime import datetime
from keras.callbacks import ModelCheckpoint
import math
checkpoint = ModelCheckpoint(filepath='mymodel.h5', 
                               verbose=1, save_best_only=True, monitor="val_acc")
callbacks = [checkpoint]

start = datetime.now()

history=model.fit_generator(
    steps_per_epoch = math.floor(len(train_set)/32),generator=train_set,validation_data = test_set,epochs=10,validation_steps=32,callbacks=callbacks ,verbose=3) # validation_data=test_set,

duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
model.evaluate(test_set)
```
